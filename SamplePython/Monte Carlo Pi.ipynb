```python
import random
def approxpi(numsamples):
    count = 0
    for i in range(0, numsamples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x**2+y**2<1):
            count+=1
    piapprox = 4.0*count/numsamples
    return piapprox
```

```python
mypi = approxpi(10000)
print(mypi)
```
