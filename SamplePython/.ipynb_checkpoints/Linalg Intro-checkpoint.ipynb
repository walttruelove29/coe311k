```python
v_add = v1+v2
v_sub = v1-v2
print("v1+v2:", v_add)
print("v1-v2:", v_sub)
```

```python
dot = np.dot(v1, v2)
print("Dot product of v1 and v2:", dot)
```

```python
mag = np.linalg.norm(v1)
print("Magnitude of v1:", mag)
```

A system of linear equations can be written in matrix form as:
Ax=b
A is the coefficient matrix
x is the unknowns vector
b is the constant vector
Given the equations:
2x+3y = 5
4x-y = 3
We can solve using numpy

```python
A = np.array([[2, 3], [4, -1]])
b = np.array([5, 3])
x = np.linalg.solve(A, b)
print("Solution (x, y):", x)
```

Finding the Eigenvalues and Eigenvectors for a matrix:

```python
A = np.array([[2, 1], [1, 2]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Inverse of a matrix
A = np.array([[4, 7], [2, 6]])
A_inv = np.linalg.inv(A)
print("Inverse of A:\n", A_inv)
identity = np.dot(A, A_inv)
print("Identity:\n", identity)
```

```python
# Define a matrix
A = np.array([[1, 2], [3, 4], [5, 6]])
# Perform SVD
U, sigma, Vt = np.linalg.svd(A)
print("U matrix:\n", U)
print("Sigma values:", sigma)
print("V^T matrix:\n", Vt)
```
