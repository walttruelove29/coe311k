```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
def approxpi(numsamples):
    count = 0
    for i in range(0, numsamples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x**2+y**2<1):
            count+=1
    piapprox = 4.0*count/numsamples
    return piapprox
```

```python
x = np.arange(10, 10000, 20)
y = []

for i in x:
    z = approxpi(i)
    error = np.abs(z-np.pi)
    y.append(error)
    
plt.plot(x, y)
plt.xlabel('Samples')
plt.ylabel('Error')
plt.title('Error as Samples Increase')
plt.show()
```
