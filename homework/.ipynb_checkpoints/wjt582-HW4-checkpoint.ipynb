```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
A = np.array([[3, -0.1, -0.2],
             [0.1, 7, -0.3],
             [0.3, -0.2, 10]], dtype=float)
b = np.array([7.85, -19.3, 71.4], dtype=float)
x0 = np.zeros_like(b)
tol1 = 10e-3
tol2 = 10e-6
tol3 = 10e-9
```

Jacobi Test:

```python
solj1, itej1 = jacobi(A, b, x0, tol1, 100)
solj2, itej2 = jacobi(A, b, x0, tol2, 100)
solj3, itej3 = jacobi(A, b, x0, tol3, 100)
print("Solution for tolerance = 10e-3:", solj1)
print("Iterations for tolerance = 10e-3:", itej1)
print("Solution for tolerance = 10e-6:", solj2)
print("Iterations for tolerance = 10e-6:", itej2)
print("Solution for tolerance = 10e-9:", solj3)
print("Iterations for tolerance = 10e-9:", itej3)
```

Of note, as the tolerance increases, the solutions become more precise and it takes more iterations for the function to reach a solution

Gauss-Seidel Test:

```python
solg1, iteg1 = gauss_seidel(A, b, x0, tol1, 100)
solg2, iteg2 = gauss_seidel(A, b, x0, tol2, 100)
solg3, iteg3 = gauss_seidel(A, b, x0, tol3, 100)
print("Solution for tolerance = 10e-3:", solg1)
print("Iterations for tolerance = 10e-3:", iteg1)
print("Solution for tolerance = 10e-6:", solg2)
print("Iterations for tolerance = 10e-6:", iteg2)
print("Solution for tolerance = 10e-9:", solg3)
print("Iterations for tolerance = 10e-9:", iteg3)
```

The same is true for the Gauss-Seidel method. Additionally, it reaches a solution in all three cases with less iterations, and it reaches the same solution with the highest two tolerances, though this may be due to truncation error in the second case

Jacobi Error Analysis:

```python
sol = np.linalg.solve(A, b)
error1 = np.abs(sol[0] - solj1[0])
error2 = np.abs(sol[1] - solj1[1])
error3 = np.abs(sol[2] - solj1[2])
jerror1 = error1+error2+error3

error1 = np.abs(sol[0] - solj2[0])
error2 = np.abs(sol[1] - solj2[1])
error3 = np.abs(sol[2] - solj2[2])
jerror2 = error1+error2+error3

error1 = np.abs(sol[0] - solj3[0])
error2 = np.abs(sol[1] - solj3[1])
error3 = np.abs(sol[2] - solj3[2])
jerror3 = error1+error2+error3

jerror = [jerror1, jerror2, jerror3]
iterations = [itej1, itej2, itej3]

plt.plot(iterations, jerror)
plt.title("Convergence for Jacobi Method")
plt.xlabel("Iterations")
plt.ylabel("Absolute Error")
plt.show()
```

```python
error1 = np.abs(sol[0] - solg1[0])
error2 = np.abs(sol[1] - solg1[1])
error3 = np.abs(sol[2] - solg1[2])
gerror1 = error1+error2+error3

error1 = np.abs(sol[0] - solg2[0])
error2 = np.abs(sol[1] - solg2[1])
error3 = np.abs(sol[2] - solg2[2])
gerror2 = error1+error2+error3

error1 = np.abs(sol[0] - solg3[0])
error2 = np.abs(sol[1] - solg3[1])
error3 = np.abs(sol[2] - solg3[2])
gerror3 = error1+error2+error3

gerror = [gerror1, gerror2, gerror3]
iterations = [iteg1, iteg2, iteg3]

plt.plot(iterations, gerror)
plt.title("Convergence for Gauss-Seidel Method")
plt.xlabel("Iterations")
plt.ylabel("Absolute Error")
plt.show()
```

In both models, the error decreased, and eventually reached zero as the computed and actual solutions converged, as the iterations increased. As for the tolerance, as the tolerance increased, more iterations were required to compute a solution. This is because the tolerance essentially sets the degree of precision of the answer. The smaller the tolerance, the closer the computed answer will be to the true one. As a result of this, it takes more iteration for each method to reach a solution. Ultimately, a smaller tolerance increases the number of iterations needed to reach a solution, while also increasing the accuracy of the computed solution.
