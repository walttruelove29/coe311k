```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def truncsin(i, x):
    sinx=0.0
    for n in range(0, i):
        term = ((-1)**n)*((x**(2*n+1))/(math.factorial(2*n+1)))
        sinx += term
    return sinx
```

```python
x = np.arange(1, 10.1, 0.1)
sin1t = truncsin(1, x)
sin2t = truncsin(2, x)
sin3t = truncsin(3, x)
sin4t = truncsin(4, x)
truesin = np.sin(x)
```

```python
t1err = np.abs(sin1t - truesin)
t2err = np.abs(sin2t - truesin)
t3err = np.abs(sin3t - truesin)
t4err = np.abs(sin4t - truesin)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, t1err)
plt.title('1 Term')
plt.xlabel('X-Values')
plt.ylabel('Error')
plt.subplot(2, 2, 2)
plt.plot(x, t2err)
plt.title('2 Terms')
plt.xlabel('X-Values')
plt.ylabel('Error')
plt.subplot(2, 2, 3)
plt.plot(x, t3err)
plt.title('3 Terms')
plt.xlabel('X-Values')
plt.ylabel('Error')
plt.subplot(2, 2, 4)
plt.plot(x, t4err)
plt.title('4 Terms')
plt.xlabel('X-Values')
plt.ylabel('Error')
plt.tight_layout()
plt.show()
```

As you add terms to the series, the error behaves in different ways. While the error decreases for smaller values of x the more terms you have, as one might expect, it increases seemingly exponentially as the value of x increases with more terms. This is likely because the additional terms allow the computed value to fluctuate far more as x increases, but there are still not enough terms being used for it to be reduced to its true value. With only one term in the series, the output value can only be greater or less than the true value by about 10, but with 2, 3, or 4 terms this possible gap grows by hundreds and thousands.
