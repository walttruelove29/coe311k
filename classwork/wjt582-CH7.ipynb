```python
import numpy as np
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

Challenge A:

```python
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
print("Jacobi:")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print("Gauss-Seidel:")
solution2, iteration2 = gauss_seidel(A, b, x0, tol, max_iterations)
print(f"Solution: {solution2}")
print(f"Iterations: {iteration2}")
```

Jacobi did not converge; Gauss-Seidel did

Challenge B:

```python
A = np.array([[1, 	1], [1, 1.0001]])
b = np.array([1, 1])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
print("Jacobi:")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print("Gauss-Seidel:")
solution2, iteration2 = gauss_seidel(A, b, x0, tol, max_iterations)
print(f"Solution: {solution2}")
print(f"Iterations: {iteration2}")
```

Both Jacobi and Gauss-Seidel Converged

Challenge C:

```python
A = np.array([[1, 	1], [1, 1.0001]])
b = np.array([1, 1.0001])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
print("Jacobi:")
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print("Gauss-Seidel:")
solution2, iteration2 = gauss_seidel(A, b, x0, tol, max_iterations)
print(f"Solution: {solution2}")
print(f"Iterations: {iteration2}")
```

Both methods failed in this instance. Introducing the .0001 in b caused the system to blow up, which implies that the matrix is ill-conditioned
