```python
import numpy as np
```

```python
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)

def diagdominant(A):
    n = len(A)
    
    for i in range(n):
        A1 = A[i][i]
        A2 = 0.0
        for j in range(n):
            if i != j:
                A2 += A[i][j]
        check = np.abs(A1)-np.abs(A2)
        if check<0:
            return print("Is the Matrix Diagonally Dominant? False \n", A)
    return print("Is the Matrix Diagonally Dominant? True \n", A)
            
```

```python
diagdominant(A)
```

This particular matrix is diagonally dominant, and if checked by hand one will reach the same result as the function
