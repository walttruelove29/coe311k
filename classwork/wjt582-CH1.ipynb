```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
plt.plot(x, np.sin(x))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x)*np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x)*np.cos(x))
plt.subplot(2, 2, 4)
plt.plot(x, np.tan(x))
```

```python
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```
