```python
import numpy as np
```

```python
piapprox = 0.0
count = 0
term = 0.0
for n in range(0,5000):
    term = (-1)**n/(2*n+1)
    piapprox+=term*4
    count+=1
    if(count == 100):
        print(np.abs(piapprox-np.pi)/np.pi)
        count = 0
```

As each term is added to the summation, the error gets smaller and smaller.
