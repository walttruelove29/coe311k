```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
def approxpi(numsamples):
    count = 0
    for i in range(0, numsamples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x**2+y**2<1):
            count+=1
    piapprox = 4.0*count/numsamples
    return piapprox
```

```python
aerror = []
rerror = []
for i in range(0, 100):
    
    estpi = approxpi(1000)
    
    err1 = np.abs(np.pi - estpi)
    aerror.append(err1)
    err2 = np.abs((np.pi-estpi)/np.pi)
    rerror.append(err2)

x = np.linspace(0, 100, 100)

plt.figure()
plt.plot(x, aerror, '-b', label='Absolute Error')
plt.title('Absolute and Relative Error vs Number of Tests')
plt.plot(x, rerror, '-k', label='Relative Error')
plt.legend()
plt.show()
```
