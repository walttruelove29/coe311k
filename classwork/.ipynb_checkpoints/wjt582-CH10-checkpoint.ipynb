```python
import numpy as np
import matplotlib.pyplot as plt
```

# On the actual assignment for the challenge it says to use random vectors for x and y from challenge 9, but in the lecture we were told to use the function plus error from challenge 9 so I went ahead and did both

Arbitrary Function + Error:

```python
X = np.arange(0, 1.01, 0.01)
error = np.random.uniform(-1, 1, len(X))
Y = 2*X+2+error
```

Least Square Fit:

```python
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Calculated Slope: {m}, Calculated Intercept: {c}")
y1 = m*X+c
```

```python
plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(X, Y, label='Data Points')
plt.plot(X, y1, label='Best Fit Line')
plt.grid(True)
plt.legend()
plt.title("Least Squares Fit")
plt.show()
```

```python
# Mean Squared Error (MSE)
mse1 = np.mean((Y - y1)**2)
print(f'Mean Squared Error: {mse1:.4f}')
# R-squared value
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - y1)**2)
r_squared1 = 1 - (ss_res / ss_total)
print(f'R-squared: {r_squared1:.4f}')
```

Polynomial Fit:

```python
degree = 3 # Degree of the polynomial
coefficients = np.polyfit(X, Y, degree)
# Print the polynomial coefficients
print("Polynomial Coefficients: ", coefficients)
# Step 3: Use the coefficients to predict y-values (polynomial function)
# Generate polynomial function based on the coefficients
polynomial = np.poly1d(coefficients)
```

```python
# Generate predicted y-values based on the fitted polynomial
y_pred = polynomial(X)
# Step 4: Plotting the results
plt.scatter(X, Y, color='blue', label='Data points') # Original data points
plt.plot(X, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
# Fitted curve
plt.title('Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Mean Squared Error (MSE)
mse2 = np.mean((Y - y_pred)**2)
print(f'Mean Squared Error: {mse2:.4f}')
# R-squared value
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - y_pred)**2)
r_squared2 = 1 - (ss_res / ss_total)
print(f'R-squared: {r_squared2:.4f}')
```

Comparison:

```python
if r_squared1 > r_squared2:
    print("The Least Squared Method provides the Best Fit for the Data")
if r_squared1 < r_squared2:
    print("Polynomial Fitting provides the Best Fit for the Data")
```

## Vandermonde Matrix:

Arbitrary Function + Error:

```python
X = np.arange(0, 1.01, 0.01)
error = np.random.uniform(-1, 1, len(X))
Y = 2*X+2+error
```

```python
# Create Vandermonde matrix for a polynomial of degree 3
degree = 3
V = np.vander(X, N=degree+1, increasing=True)
```

```python
# V.T * V
VT_V = np.dot(V.T, V)
# V.T * y
VT_y = np.dot(V.T, Y)
# Solve for a (coefficients)
coefficients = np.linalg.solve(VT_V, VT_y)
print(coefficients)
```

```python
polynomial = np.poly1d(coefficients[::-1])

y_pred = polynomial(X)

plt.scatter(X, Y, color='blue', label='Data points') 
plt.plot(X, y_pred, color='red', label=f'Vandermonde Polynomial fit of degree {degree}')

plt.title('Polynomial Fit with Vandermonde Matrix')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Mean Squared Error (MSE)
mse = np.mean((Y - y_pred)**2)
print(f'Mean Squared Error: {mse:.4f}')
# R-squared value
ss_total = np.sum((Y - np.mean(Y))**2)
ss_res = np.sum((Y - y_pred)**2)
r_squared = 1 - (ss_res / ss_total)
print(f'R-squared: {r_squared:.4f}')
```
